```python
score = input ("Enter your grade: ")

if int(score) > 97 and int(score) < 100:
    print("1.0")
elif int(score) > 93 and int(score) < 97:
    print("1.25")
elif int(score) > 89 and int(score) < 93:
    print("1.5")
elif int(score) > 85 and int(score) < 89:
    print ("1.75")
elif int(score) > 81 and int(score) < 85:
    print("2.0")
elif int(score) > 77 and int(score) < 81:
    print("2.25")
elif int(score) > 74 and int(score) < 77:
    print ("2.5")
else:
    print ("3.0")
```
