Exercise 1
Investigate on the difference between is and == given the following variables.

a = 2
b = 2.0
c = a

```python
a = 2       # == checks if the values of two things are the same.
b = 2.0      # is for checking object identity (whether two variables refer to the same object).
c = a

print (a == 2)    # true because the values are the same
print(a is b)     #false because different types( integer and float)
print(a == c)     #true because the value of a and c are equal
print(a is c)     #true becase the value are equal and the same type
```

Challenge 1
Write a program which repeatedly reads numbers until the user enters "done". Once "done" is entered, print out the total, count, and average of the numbers. If the user enters anything other than a number, detect their mistake using try and except and print an error message and skip to the next number.

Enter a number: 4
Enter a number: 5
Enter a number: bad data
Invalid input
Enter a number: 7
Enter a number: done
16 3 5.333333333333333

```python
total = 0
count = 0

while True:
        try:
          input_number = input("Enter a number: ")
          if input_number == 'done':
             break
          number = int(input_number)
          total += number
          count += 1

        except ValueError:
          print("Invalid input")

if count > 0:
          ave = total / count
          print(total, count, ave)
else:
          print("No number entered.")
```

Challenge 2
Write another program that prompts for a list of numbers as above and at the end prints out both the maximum and minimum of the numbers instead of the average.

```python
number = []

while True:
    try:
      number_input = input("Enter a number: ")
      if number_input == 'done':
        break
      num = float(number_input)
      number.append(num)
    except ValueError:
      print("Invalid input")

if number:
    maximum = max(number)
    minimum = min(number)
    print("Maximum:", maximum)
    print("Minimum:", minimum)

else:
    print("No numbers entered.")

```
