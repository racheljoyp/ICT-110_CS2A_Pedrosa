Exercise 2

Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours. Example:

```python
Enter Hours: 45
Enter Rate: 10
Pay: 475.0
```

answer:

```python
hours = float (input(" Enter Hours: "))
rate = float (input ("Enter rate: "))1

if hours > 40:
   regular = 40 * rate
   overtime = (hours - 40) * rate * 1.5
   pay = regular + overtime

else :
  pay = hours * rate

print ("Pay: ", pay)
```

CHALLENGE
Write a program to prompt for a score between 0 and 100 and print a grade using the following table:
Score           Grade
[ 97, 100]      1.0
[ 93,  97)        1.25
[ 89,  93)        1.5
[ 85,  89)        1.75
[ 81,  85)        2.0
[ 77,  81)        2.25
[ 74,  77)        2.5
[ 0,   74)        3.0

```python
score = input ("Enter score: ")

if int(score) >= 97 and int(score) <= 100:
    grade = 1.0
elif int(score) >= 93 and int(score) < 97:
     grade = 1.25
elif int(score) >= 89 and int(score) < 93:
     grade = 1.5
elif int(score) >= 85 and int(score) < 89:
    grade = 1.75
elif int(score) >= 81 and int(score) < 85:
     grade = 2.0
elif int(score) >= 77 and int(score) < 81:
     grade = 2.2
elif int(score) >= 74 and int(score) < 77:
     grade = 2.5
else:
      grade = 3.0

print("Your grade is: ",grade)
```
